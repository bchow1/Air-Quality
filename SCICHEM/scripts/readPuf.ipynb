**Analysis of puff output using pandas**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv file created by readpuf utility
pufCsvFile = '../output/no2-2005-mc/no2-2005-mc-puf.csv'
```

```python
# Read puff data into data frame and print the columns.
puffDf = pd.read_csv(pufCsvFile,skiprows=2)

# Drop last blank column
puffDf.drop(puffDf.columns[len(puffDf.columns)-1], axis=1, inplace=True)
print puffDf.columns
```

```python
# Group by output time
puffTime = puffDf.groupby(['TIME'],as_index=False)
```

```python
# No. of puffs
print puffTime['IPUF'].max().head(5)
puffTime['IPUF'].max().plot(x = "TIME",y = "IPUF")
plt.title('No. of Puffs')
plt.show()
```

```python
# Total mass
print puffTime['C'].sum().head(5)
puffTime['C'].sum().plot(x = "TIME",y = "C")
plt.title('Total Mass')
plt.show()
```

```python
# C_NO2
print puffTime['C_NO2'].sum().head(5)
puffTime['C_NO2'].sum().plot(x = "TIME",y = "C_NO2")
plt.title('NO2')
plt.show()
```
